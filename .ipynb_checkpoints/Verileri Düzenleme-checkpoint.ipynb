```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```

```python
print("Top Absolute Correlations")
print(get_top_abs_correlations(new.loc[:, "BESLENEN_HAYVAN_TÜRLERİ_BÜYÜKBAŞ" : "İYİ_TARIM_UYGULAMALARI"], 100))
```

```python
for colm in ["SULAK_ALAN_MİKTARI", "TRAKTÖR_SAYISI", "BİREYSEL_ARAZİ_ORTALAMASI"]:
    new[colm] = new[colm].astype('int')
    new[colm] = (new[colm] - new[colm].min())/(new[colm].max() - new[colm].min())
```

```python
target = new.pop("İYİ_TARIM_UYGULAMALARI")
data = new.loc[:, "BESLENEN_HAYVAN_TÜRLERİ_BÜYÜKBAŞ" : "TRAKTÖR_SAYISI"]
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf

target = np.array(target).reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.33, random_state=42)

X_train, X_test = tf.convert_to_tensor(np.array(X_train)), tf.convert_to_tensor(np.array(X_test))
y_train, y_test = tf.convert_to_tensor(np.array(y_train)), tf.convert_to_tensor(np.array(y_test))
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
BATCH_SIZE = 5

def get_basic_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(278, activation='relu'),
        tf.keras.layers.Dense(140, activation='relu'),
        tf.keras.layers.Dense(70, activation='relu'),
        tf.keras.layers.Dense(1)
    ])

    model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
    return model
```

```python
model = get_basic_model()
model.fit(X_train, y_train, epochs=15, batch_size=BATCH_SIZE)
```

```python
model.evaluate(X_test,  y_test, verbose=2)
```
