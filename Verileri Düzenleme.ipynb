```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from functools import reduce
import glob
import os

pd.set_option('display.max_rows', None)
```

```python
path = r'C:\Users\Dilemre\GitHub\donem\data'                    
all_files = glob.glob(os.path.join(path, "*.csv"))    
dfList = [pd.read_csv(f,  sep=";", encoding='iso-8859-9', usecols=range(3), lineterminator='\n', skip_blank_lines=True) for f in all_files]
data = reduce(lambda x, y: pd.merge(x, y), dfList)
```

```python
data.to_csv("raw_data.csv")
```

```python
data = data.replace({'\r':''}, regex=True)
data = data.replace({' Adet':''}, regex=True)
data = data.replace({' Dekar':''}, regex=True)
data = data.replace({'%':''}, regex=True)
data = data.replace({'': np.nan}, regex=True)

data.columns = data.columns.str.replace('\r', '')
data.columns = data.columns.str.replace(' ', '_')
```

```python
data = data.dropna().reset_index()
```

```python
convert_list = ["BESLENEN_HAYVAN_TÜRLERİ", "SULAMA_SUYU_KAYNAĞI_YÖNETİMİ", "HAYVAN_BESLEMEDE_KULLANILAN_ÜRÜNLER",
                "SULANABİLİR_ALANDA_YETİŞTİRİLEN_ÜRÜNLER", "TARIMSAL_SULAMA_SUYU_KAYNAĞI", 
                "DOĞRUDAN_SATIN_ALINABİLECEK_ÜRÜNLER", "MEŞHUR_ÜRÜNLER", "EN_ÇOK_YETİŞTİRİLEN_ÜRÜNLER", 
                "ENDEMİK_(SADECE_BU_BÖLGEDE_YETİŞEN)_BİTKİLER", "TALEP_EDİLEN_TARIMSAL_EĞİTİMLER", 
                "TIBBİ_VE_AROMATİK_BİTKİ_YETİŞTİRİCİLİĞİ", "ALTERNATİF_YENİ_ÜRÜN_ÇEŞİTLERİ_ÜRETİMİ"]

other_list = ["İLÇE_ADI", "MAHALLE_ADI", "ARICILIK", "TARIMSAL_SULAMA_SUYU", "SULAK_ALAN_MİKTARI", "TRAKTÖR_SAYISI", 
              "BİREYSEL_ARAZİ_ORTALAMASI", "İYİ_TARIM_UYGULAMALARI"]

new = data.loc[:, ("İLÇE_ADI", "MAHALLE_ADI")].copy().astype('category')

for colm in convert_list:
    temp = np.unique(', '.join(data[colm]).split(', '))
    
    for ctgr in temp:
        new[colm+"_"+ctgr] = data[colm].str.contains(ctgr).astype('int')
        
new = pd.merge(new, data[other_list])
```

```python
new.tail()
```

```python
new.columns = new.columns.str.replace(' ', '_')
```

```python
for colm in ["ARICILIK", "TARIMSAL_SULAMA_SUYU", "İYİ_TARIM_UYGULAMALARI"]:
    new[colm]  = [1 if (i == "EVET" or i == "VAR") else 0 for i in new[colm]]
```

```python
new.tail()
```

```python
for colm in ["SULAK_ALAN_MİKTARI", "TRAKTÖR_SAYISI", "BİREYSEL_ARAZİ_ORTALAMASI"]:
    new[colm] = new[colm].astype('int')
    new[colm] = (new[colm] - new[colm].min())/(new[colm].max() - new[colm].min())
```
